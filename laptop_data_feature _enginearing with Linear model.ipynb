```python
X = df.drop(columns=['Price'])
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.20, random_state=20)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
make_column=make_column_transformer(
     (OneHotEncoder(sparse_output = False,handle_unknown='ignore',categories='auto'),['Company', 'TypeName']),
     (StandardScaler(),['Inches', 'Ram', 'Weight', 'Price', 'IPS Panel', 'Touchscreen',
       'price per inches', 'Intel Core i7', 'Intel Core i5',
       'other Intel Processer', 'AMD Processer', 'Intel_Graphics',
       'AMD_Graphics', 'Nvidia_Graphics', 'Windoes_opsys', 'MacOs_opsys',
       'other_opsys', 'SSD', 'HDD', 'Flash Storage', 'Hybrid']),
     remainder='passthrough')
```

```python
make_column
```

```python
regressor=LinearRegression()
```

```python
pipe=make_pipeline(make_column,regressor)
```

```python
pipe.fit(X_train,y_train)
```

```python
pred = pipe.predict(X_test)
```

```python
r2_score(y_test,pred)
```
